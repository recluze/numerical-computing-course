```python
from pandas import read_csv
```

```python
dataframe = read_csv("data/longley.csv", header=None)
data = dataframe.values
print(data)
```

```python
x, y = data[:, 4], data[:, -1]
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(x, y)
plt.show()
```

```python
def objective_1(x, a, b): 
    return a * x + b
```

```python
from scipy.optimize import curve_fit 
```

```python
popt, _ = curve_fit(objective_1, x, y)
```

```python
a, b = popt
print('y = %.5f * x + %.5f' % (a, b))
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(x, y)
x_line = np.arange(min(x), max(x), 0.1)
y_1 = objective_1(x_line, a, b)
plt.plot(x_line, y_1, '--')
plt.show()
```

```python
def objective_2(x, a, b, c): 
    return a * x + b * x**2 + c
```

```python
popt, _ = curve_fit(objective_2, x, y)
a, b, c = popt

print('y = %.5f * x + %.5f * xˆ2 + %.5f' % (a, b, c))


plt.figure(figsize = (8, 6))
plt.scatter(x, y)
x_line = np.arange(min(x), max(x), 0.1)
y_2 = objective_2(x_line, a, b, c)
plt.plot(x_line, y_2, '--')
plt.show()
```

```python
def objective_5(x, a, b, c, d, e, f):
    return (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + f
```

```python
popt, _ = curve_fit(objective_5, x, y)
a, b, c, d, e, f = popt


plt.figure(figsize = (8, 6))
plt.scatter(x, y)
x_line = np.arange(min(x), max(x), 0.1)
y_5 = objective_5(x_line, a, b, c, d, e, f)
plt.plot(x_line, y_5, '--')
plt.show()
```

```python
from numpy import sin 

def objective_sin(x, a, b, c, d):
    return a * sin(b - x) + c * x**2 + d
```

```python
popt, _ = curve_fit(objective_sin, x, y)
a, b, c, d = popt


plt.figure(figsize = (8, 6))
plt.scatter(x, y)
x_line = np.arange(min(x), max(x), 0.1)
y_sin = objective_sin(x_line, a, b, c, d)
plt.plot(x_line, y_sin, '--')
plt.show()
```
